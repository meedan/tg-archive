```python
import sqlite3
import pandas as pd

# Replace 'your_database_file.sqlite' with the name of your SQLite database file
database_file = '/Users/caoyujia/Desktop/tg-archive/data/muthanapress84/data.sqlite'

# Create a connection to the SQLite database file
connection = sqlite3.connect(database_file)

# Create a cursor object to interact with the database
cursor = connection.cursor()

# Example query to read data from a table in the database
cursor.execute("SELECT * FROM messages")

# Fetch the results and store them in a Pandas DataFrame
data = cursor.fetchall()
df = pd.DataFrame(data)

# Display the DataFrame in the Jupyter Notebook
df
```

```python
# You can now work with the 'data' variable, which contains the results of the query

# Close the cursor and connection when you're done
cursor.close()
connection.close()
```
